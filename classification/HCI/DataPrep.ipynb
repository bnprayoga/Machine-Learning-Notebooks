### Dealing With Missing Values

```python
# check dataset missing values
def check_null(df, ascending = False):
    # create series of null columns and null qty
    series = df.isnull().sum()

    # take only columns that has null values
    series = series[series > 0]

    # calculate percent
    percent = 100 * series / df.shape[0]

    # check dtype
    coldtype = df[series.index].dtypes

    # create dataframe
    result = pd.DataFrame({
        'QTY' : series, 'Percent' : percent, 'dtypes' : coldtype
    })

    if ascending != None:
        result.sort_values(by = 'QTY', ascending = ascending, inplace = True)

    return result
```

```python
# check null values
column_null_on_train = check_null(application_train_df, ascending = False)
column_null_on_train
```

```python
# check null values that less than 5 percent
column_null_on_train[column_null_on_train['Percent'] <= 10]
```

```python
# there's a lot of missing values, most of them are above 10% from total row within applicant data
# we will drop all features that has nan more than 10% since it cannot give reliable informations and treat these features by filling it
# will only introduce noises to our feed data

# take all dropped columns
dropped_features = column_null_on_train[column_null_on_train['Percent'] > 10].index.to_list()

# drop all features that has missing values more than 10%
application_train_cleaned = application_train_df.drop(columns = dropped_features)
application_train_cleaned.head(3)
```

```python
# filling NAME_TYPE_SUITE with it's mode
application_train_cleaned.loc[application_train_cleaned.NAME_TYPE_SUITE.isna(), 'NAME_TYPE_SUITE'] = application_train_cleaned.loc[:, 'NAME_TYPE_SUITE'].mode()
```

```python
# filling all missing values with 0
application_train_cleaned.fillna(0, inplace=True)
```

```python
bureau_info = bureau_active_info_df.merge(bureau_balance_status_info, on = 'SK_ID_BUREAU', how = 'left')
bureau_info = bureau_info.groupby(by = 'SK_ID_CURR').sum().reset_index()
bureau_info.drop(columns = 'SK_ID_BUREAU', inplace=True)
bureau_info.fillna(0, inplace=True)
bureau_info
```
